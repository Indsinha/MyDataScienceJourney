** Import pandas as pd.**

** Read Salaries.csv as a dataframe called sal.**

```python
sal= pd.read_csv("Salaries - Copy.csv")
```

** Check the head of the DataFrame. **

```python
sal.head()
```

** Use the .info() method to find out how many entries there are.**

```python
sal.info()
```

**What is the average BasePay ?**

```python
sal['BasePay'].mean()
```

** What is the highest amount of OvertimePay in the dataset ? **

```python
sal['OvertimePay'].max()
```

** What is the job title of  JOSEPH DRISCOLL ? Note: Use all caps, otherwise you may get an answer that doesn't match up (there is also a lowercase Joseph Driscoll). **

```python
job_title = sal[sal['EmployeeName'] == 'JOSEPH DRISCOLL']['JobTitle']
print(job_title)
```

** How much does JOSEPH DRISCOLL make (including benefits)? **

```python
totpb = sal[sal['EmployeeName'] == 'JOSEPH DRISCOLL']['TotalPayBenefits']
print(totpb)
```

** What is the name of highest paid person (including benefits)?**

```python
highest_paid = sal.loc[sal['TotalPayBenefits'].idxmax()]
print(highest_paid['EmployeeName'])
```

** What is the name of lowest paid person (including benefits)? Do you notice something strange about how much he or she is paid?**

```python
lowest_paid = sal.loc[sal['TotalPayBenefits'].idxmin()]
print(lowest_paid['EmployeeName'])
```

** What was the average (mean) BasePay of all employees per year? (2011-2014) ? **

```python
average_basepay_per_year = sal.groupby('Year')['BasePay'].mean()
print(average_basepay_per_year)
```

** How many unique job titles are there? **

```python
sal['JobTitle'].nunique()
```

** What are the top 5 most common jobs? **

```python
top_5_common_jobs = sal['JobTitle'].value_counts().head(5)
print(top_5_common_jobs)
```

** How many Job Titles were represented by only one person in 2013? (e.g. Job Titles with only one occurence in 2013?) **

```python
job_titles_2013 = sal[sal['Year'] == 2013]['JobTitle']
unique_job_titles_2013 = job_titles_2013.value_counts()
one_occurrence_job_titles = unique_job_titles_2013[unique_job_titles_2013 == 1]
print(one_occurrence_job_titles)
```

** How many people have the word Chief in their job title? (This is pretty tricky) **

```python
chief_count = sal[sal['JobTitle'].str.contains('Chief', case=False)].shape[0]
print(chief_count)
```

** Bonus: Is there a correlation between length of the Job Title string and Salary? **

```python
# Create a new column for the length of the Job Title
sal['title_len'] = sal['JobTitle'].str.len()

# Calculate the correlation between the length of Job Title and Total Pay Benefits
correlation = sal['title_len'].corr(sal['TotalPayBenefits'])
print(correlation)
```

We observe that there is a weak -ve correlation therefore we can conclude that there is no significant relation between len of job title and salary
