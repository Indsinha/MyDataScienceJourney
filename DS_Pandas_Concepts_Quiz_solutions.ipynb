Question:
Exercise: Write a function that takes the DataFrame df and returns the percentage of total Credit Card Outstanding (k$) contributed by each Gender.

```python
def get_outstanding_percentage_by_gender(df):
    total_outstanding = df['Credit Card Outstanding (k$)'].sum()
    gender_outstanding = df.groupby('Gender')['Credit Card Outstanding (k$)'].sum()
    return (gender_outstanding / total_outstanding * 100).to_dict()

get_outstanding_percentage_by_gender(df)
```

Question:
Exercise: Write a function that takes the DataFrame df and returns a DataFrame with an additional column Cumulative Charges that shows the cumulative sum of Interest Levied (k$) and Late Payment Charges (k$) up to that row.

```python
def add_cumulative_charges(df):
    df['Cumulative Charges'] = (df['Interest Levied (k$)'] + df['Late Payment Charges (k$)']).cumsum()
    return df
add_cumulative_charges(df)
```
