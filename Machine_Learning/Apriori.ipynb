```python
from apyori import apriori
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data= pd.read_csv('./TV Shows.csv', header= None)
data
```

```python
# Stucture the data
# transaction[[Cobra Kai,Lupin,12 Monkeys,Sherlock],...,9689]
transaction=[]
for i in range(0,len(data)):
    transaction.append([str(data.values[i,j])for j in range(0,len(data.columns))])
```

```python
transaction
```

```python
transaction1=list(data.values[0:len(data)])
transaction1
```

```python

# Creating the rules using apriori

# recommendation will run/update for 7 daysÂ 

# i want 3 people should watch a movie from the total entries, min support:3*7/9690= 0.002167

# confidence 0.2 = 20%
rules=apriori(transaction,min_support=0.0021,min_confidence=0.2, min_lift = 3, max_length=2)
```

```python
rules
```

```python
result = list  (rules)
```

```python
result
```

```python
# creating a function to extract evry informationÂ 
# in the form of Dataframe
def inspect(results):
    lhs= [tuple(result[2][0][0])[0] for result in results]
    rhs= [tuple(result[2][0][1])[0] for result in results]
    support= [result[1] for result in results]
    confidence= [result[2][0][2] for result in results]
    lifts= [result[2][0][3] for result in results]
    return list(zip(lhs,rhs,support,confidence,lifts))
```
