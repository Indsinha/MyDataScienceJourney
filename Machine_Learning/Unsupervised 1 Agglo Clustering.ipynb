```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars = pd.read_csv('cars_clus.csv')
cars.head()
```

```python
cars_cl= cars.drop(columns=['manufact','model'])
cars_cl.head()
```

```python
sns.scatterplot(cars_cl['sales'], cars_cl['mpg'])
```

```python
# Normalizig the values

from sklearn.preprocessing import MinMaxScaler

scale=MinMaxScaler()

cars_scaled= scale.fit_transform(cars_cl)
cars_scaled= pd.DataFrame(cars_scaled, columns=cars_cl.columns)

cars_scaled
```

```python
from scipy.cluster import hierarchy
dendogram= hierarchy.dendrogram(hierarchy.linkage(cars_scaled,method='ward'))
```

```python
# Agglomerative approach
from sklearn.cluster import AgglomerativeClustering
hc= AgglomerativeClustering(n_clusters=2,affinity='euclidean',linkage='ward')
```

```python
clusters=hc.fit_predict(cars_scaled)
```

```python
cars['Clusters']= clusters # adding new col in Df
```

```python
cars.head()
```

```python
sns.scatterplot(cars['sales'],cars['mpg'], hue= cars['Clusters'])
```
